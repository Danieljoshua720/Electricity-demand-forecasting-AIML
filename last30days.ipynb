```python
# Convert datetime
df['datetime'] = pd.to_datetime(df['datetime'])

# Sort by time (VERY IMPORTANT)
df = df.sort_values('datetime').reset_index(drop=True)

# Keep ONLY target column for now (avoid scaler mistakes)
data = df[['Power demand']].values
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
```

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
n_past = 168        # 7 days history
n_future = 720      # last 30 days (hourly)
```

```python
train_data = data_scaled[:-n_future]
test_data  = data_scaled[-n_future:]
```

```python
X_train, y_train = [], []

for i in range(n_past, len(train_data)):
    X_train.append(train_data[i-n_past:i])
    y_train.append(train_data[i])

X_train = np.array(X_train)
y_train = np.array(y_train)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
model = Sequential([
    LSTM(64, input_shape=(n_past, 1)),
    Dense(1)
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse'
)

model.summary()
```

```python
model.fit(
    X_train,
    y_train,
    epochs=3,
    batch_size=1024,
    validation_split=0.1,
    verbose=1
)
```

```python
X_test = []

start = len(data_scaled) - n_future - n_past
end   = len(data_scaled) - n_past

for i in range(start, end):
    X_test.append(data_scaled[i:i+n_past])

X_test = np.array(X_test)
```

```python
X_test.shape
```

```python
y_pred_scaled = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred_scaled)
```

```python
y_true = scaler.inverse_transform(
    data_scaled[-n_future:]
)
```

```python
hours = np.arange(1, n_future + 1)

plt.figure(figsize=(14,5))
plt.plot(hours, y_true, label='Actual Last 30 Days', color='green')
plt.plot(hours, y_pred, label='Predicted Last 30 Days', color='red')
plt.xlabel('Hours (Last 30 Days)')
plt.ylabel('Power Demand')
plt.title('Actual vs Predicted â€“ Last 30 Days')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

MAE  = mean_absolute_error(y_true, y_pred)
MSE  = mean_squared_error(y_true, y_pred)
RMSE = np.sqrt(MSE)
R2   = r2_score(y_true, y_pred)

print("MAE  :", MAE)
print("MSE  :", MSE)
print("RMSE :", RMSE)
print("RÂ²   :", R2)
```

```python
model_stacked = Sequential([
    LSTM(64, return_sequences=True, input_shape=(n_past, 1)),
    LSTM(32),
    Dense(1)
])

model_stacked.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse'
)

model_stacked.summary()
```
